```python
# 字符900，字符648同理，参见评论区

from collections import Counter
train_df['3750'] = train_df['text'].apply(lambda x: list(x.split(' ')).count('3750'))
```

```python
train_df['3750'].sum()
```

```python
all_lines = ' '.join(train_df['text'])
total_cnt = Counter(all_lines.split(' '))
total_cnt = sorted(total_cnt.items(), key=lambda d: d[1], reverse=True)
total_cnt[: 10]
```

```python
total_cnt[: 20]
```

```python
class_group =  train_df.groupby(train_df['label'])
```

```python
class_group.size()
```

```python
unique_cnt = dict()

for label, news in class_group:
    all_lines = ' '.join(news['text'])
    class_cnt = Counter(all_lines.split(' '))
    class_cnt = sorted(class_cnt.items(), key=lambda d: d[1], reverse=True)
    unique_cnt[label] = class_cnt[: 50]
    
```

```python
unique_cnt[0][: 20]
```

```python
top_20_in_all = []
for i in range(20):
    top_20_in_all.append(total_cnt[i][0])
top_20_in_all
```

```python
top_20_unique = dict()

for i in range(14):
    cur_top_20 = []
    cur_cnt = unique_cnt[i]
    j = 0
    while j < len(cur_cnt) and len(cur_top_20) < 20:
        if cur_cnt[j][0] not in top_20_in_all:
            cur_top_20.append(cur_cnt[j])
        j += 1
    top_20_unique[i] = cur_top_20
```

```python
for k, v in top_20_unique.items():
    print(k, v)
```
